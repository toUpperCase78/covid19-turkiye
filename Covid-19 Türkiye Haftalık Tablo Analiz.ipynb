```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid19haftalik = pd.read_csv('covid_19_Turkiye_Haftalik_Tablo.csv')
```

## Covid-19 Haftalık Tablolar - Test Sayıları

```python
testSayilar = covid19haftalik['Test']
plt.style.use('seaborn')
plt.rcParams['axes.facecolor'] = '#ffffff'
plt.rcParams['axes.edgecolor'] = '#eeeeee'
plt.figure(figsize=(14,12))
plt.axis([testSayilar.min()*0.90, testSayilar.max()*1.03, -0.6, 32.6])
plt.barh(covid19haftalik['Tarih'], testSayilar, color='orange')
for i in range(len(covid19haftalik)):
    plt.text(testSayilar.iloc[i]-130000, i-0.18, testSayilar.iloc[i], fontsize=13, fontweight='bold', color='#734905')
plt.title('Covid-19 Haftalık Tablolar - Test Sayıları', fontsize=24, fontweight='bold')
plt.xlabel('Test Sayısı', fontsize=18, fontweight='bold')
plt.ylabel('Haftalık Tarihler', fontsize=18, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=11)
plt.grid(visible=True, color='#cfcfcf')
plt.show()
```

## Covid-19 Haftalık Tablolar - Hastalarda Zatürre Oranı

```python
hastaZaturre = covid19haftalik['Hastalarda Zatürre(%)']
plt.figure(figsize=(15.5,7))
plt.axis([-0.35, 32.3, hastaZaturre.min()-0.06, hastaZaturre.max()+0.1])
plt.plot(covid19haftalik['Tarih'], hastaZaturre, marker='s', markersize=12, color='#9a3bbf', linestyle='-.')
for i in range(len(covid19haftalik)):
    if i in [2, 5, 18, 19, 21, 29, 30, 31, 32]:   h = -0.05
    else:   h = 0.025
    plt.text(i-0.4, hastaZaturre.iloc[i]+h, "{:.2f}".format(hastaZaturre.iloc[i]), color='#9a3bbf', fontsize=12)
plt.title('Covid-19 Haftalık Tablolar - Hastalarda Zatürre Oranı', fontsize=24, fontweight='bold')
plt.xlabel('Haftalık Tarihler', fontsize=18, fontweight='bold')
plt.ylabel('Hastalarda Zatürre Oranları (%)', fontsize=18, fontweight='bold')
plt.xticks(fontsize=11, rotation=80)
plt.yticks(fontsize=14)
plt.grid(visible=True, color='#cfcfcf')
plt.show()
```
